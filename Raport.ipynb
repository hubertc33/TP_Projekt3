# DFT i IDFT


Do argumentu fukncji obliczającej DFT lub IDFT należy podać strukture zawierająca wektory wartości i czasu
Do obliczenia DFT korzystamy z fukncji:

s.dft(struktura)

Do obliczenie IDFT:

s.idft(struktura)

Działanie DFT i IDFT pokarzemy dla sygnałów sin, cos ,prostokątny i piłoksztaltny o częstotliwości 10Hz i 100Hz oraz dla sygnału audio
poprzez wykonanie na sygnale DFT, a następnie IDFT

- Przykład dla sinusa o częstotliwości 10Hz

```python
import scikit_build_example as s

a="sin"
b=s.dft(s.gen_signal(a,10))
s.show_plot(b)
s.show_plot(s.idft(b))
```
